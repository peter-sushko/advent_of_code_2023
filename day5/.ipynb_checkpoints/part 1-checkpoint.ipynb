```python
def find_seeds():
    with open('data.txt', 'r', encoding='utf-8') as file:
        first_line = file.readline()
        seeds = first_line[:-1].split(' ') # the -1 gets rid of the \n character
    return seeds[1:] #first one is "seeds:" so i drop it
```

```python
def find_start_end_of_mapping(start_keyword):
    """Given a keyword, outputs start row and end row where the mapping is located.
    
    
    
    """
   
    with open('data.txt', 'r', encoding='utf-8') as file:
        start_index = 0
        end_index = 0

        for index,line in enumerate(file):
            if line.strip().lower().startswith(start_keyword):
                start_index = end_index = index

            elif start_index != 0 and not line.strip():  # Check for a blank line after "soil"
                end_index = index
                break  
    print(start_index, end_index)
    return (start_index, end_index)
```

```python
def find_mapping(keyword):
    """Given a keyword, outputs start row and end row where the mapping is located.

    
    """
    
    with open('data.txt', 'r', encoding='utf-8') as file:
        for line in file:
            if not line.strip().lower().startswith(keyword):
                pass
            print(line)
            
        return
```
